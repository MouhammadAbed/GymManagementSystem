# [dbo].[fnCheckTrainerAssignmentStatus]

```sql
/****** Object:  UserDefinedFunction [dbo].[fnCheckTrainerAssignmentStatus]    Script Date: 8/1/2025 5:13:56 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE FUNCTION [dbo].[fnCheckTrainerAssignmentStatus]
(
    @TrainerId INT,
    @CategoryId INT
)
RETURNS INT
AS
BEGIN
    DECLARE @status INT = 0;

    IF EXISTS (
        SELECT 1 FROM TrainerCategoryAssignments 
        WHERE TrainerId = @TrainerId AND CategoryId = @CategoryId AND IsDelete = 0
    )
        SET @status = 1;
    ELSE IF EXISTS (
        SELECT 1 FROM TrainerCategoryAssignments 
        WHERE TrainerId = @TrainerId AND CategoryId = @CategoryId AND IsDelete = 1
    )
        SET @status = 2;

    RETURN @status;
END
GO
```

# [dbo].[GetAssignmentId]

```sql
/****** Object:  UserDefinedFunction [dbo].[GetAssignmentId]    Script Date: 8/1/2025 5:13:56 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create Function [dbo].[GetAssignmentId]
(
	@categoryName varchar,
	@FullName int
)
RETURNS INT
AS
BEGIN
DECLARE @AssignmentId int;
SELECT @AssignmentId= TrainerCategoryAssignments.AssignmentId from TrainerCategoryAssignments
inner join Trainers on Trainers.trainer_id = TrainerCategoryAssignments.TrainerId
inner join People on People.personid=Trainers.person_Id
inner join Categories on categories.category_id=TrainerCategoryAssignments.CategoryId
where Categories.category_name=@categoryName and People.first_name=@FullName
RETURN @AssignmentId
END;
GO
```

# [dbo].[SubscribersFullInfo]

```sql
/****** Object:  View [dbo].[SubscribersFullInfo]    Script Date: 8/1/2025 5:13:56 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE VIEW [dbo].[SubscribersFullInfo] AS        
SELECT dbo.Subscriptions.subscription_id AS [SUB ID], LTRIM(RTRIM(      
        MemberPeople.first_name +      
        CASE       
            WHEN MemberPeople.second_name IS NULL OR MemberPeople.second_name = '' THEN ''      
            ELSE ' ' + MemberPeople.second_name      
        END +      
        ' ' + MemberPeople.last_name      
    )) AS [MEMBER NAME],    
	LTRIM(RTRIM(      
        TrainerPeople.first_name +      
        CASE       
            WHEN TrainerPeople.second_name IS NULL OR TrainerPeople.second_name = '' THEN ''      
            ELSE ' ' + TrainerPeople.second_name      
        END +      
        ' ' + TrainerPeople.last_name      
    )) AS [TRAINER NAME],
dbo.Categories.category_name AS [CATEGORY],        
CAST(dbo.Subscriptions.subscriptions_start_date AS DATE) AS [START DATE],        
CAST( dbo.Subscriptions.subscriptions_end_date AS DATE) AS [END DATE],        
CASE   
    WHEN CAST(Subscriptions.subscriptions_end_date AS DATE) >= CAST(GETDATE() AS DATE)  
    THEN 'Active'  
    ELSE 'InActive'  
END AS [STATUS],        
dbo.Subscriptions.fees AS [FEES], dbo.Subscriptions.discount, dbo.Subscriptions.final_fees as [FINAL FEES],Users.userName    
FROM dbo.Subscriptions INNER JOIN        
dbo.Applications ON dbo.Applications.application_id = dbo.Subscriptions.Application_id INNER JOIN        
dbo.members ON dbo.members.memberId = dbo.Applications.member_id INNER JOIN        
dbo.People AS MemberPeople ON MemberPeople.personid = dbo.members.person_Id INNER JOIN        
dbo.TrainerCategoryAssignments ON dbo.TrainerCategoryAssignments.AssignmentId = dbo.Applications.assignment_id INNER JOIN        
dbo.Trainers ON dbo.Trainers.trainer_id = dbo.TrainerCategoryAssignments.TrainerId INNER JOIN        
dbo.People AS TrainerPeople ON TrainerPeople.personid = dbo.Trainers.person_Id INNER JOIN        
dbo.Categories ON dbo.Categories.category_id = dbo.TrainerCategoryAssignments.CategoryId    
INNER JOIN Users ON Users.userid = Subscriptions.Register_via_user    
where (dbo.Subscriptions.isDelete = 0) AND (dbo.Subscriptions.isRefund = 0)
GO
```

# [dbo].[prDeleteTrainerAndAssignments]

```sql
/****** Object:  StoredProcedure [dbo].[prDeleteTrainerAndAssignments]    Script Date: 8/1/2025 5:13:56 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[prDeleteTrainerAndAssignments]
    @PersonId INT
AS
BEGIN
    BEGIN TRY
        BEGIN TRANSACTION;

        DECLARE @TrainerId INT;

        -- Get trainer_id based on person_id
        SELECT @TrainerId = trainer_id
        FROM trainers
        WHERE person_id = @PersonId;

        -- If trainer not found, exit
        IF @TrainerId IS NULL
        BEGIN
            ROLLBACK TRANSACTION;
            RETURN;
        END

        -- Soft delete the trainer
        UPDATE trainers
        SET isDeleted = 1
        WHERE trainer_id = @TrainerId;

        -- Soft delete trainer assignments
        UPDATE TrainerCategoryAssignments
        SET IsDelete = 1
        WHERE TrainerId = @TrainerId;

        COMMIT TRANSACTION;
    END TRY
    BEGIN CATCH
        ROLLBACK TRANSACTION;

    END CATCH
END
GO
```

# [dbo].[prSubscription_AddSubscriptionAndManageApplication]

```sql
/****** Object:  StoredProcedure [dbo].[prSubscription_AddSubscriptionAndManageApplication]    Script Date: 8/1/2025 5:13:56 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[prSubscription_AddSubscriptionAndManageApplication]
    @PersonId INT,          
    @assigment_Id  INT,          
    @subscription_start_date DATETIME,          
    @subscription_end_date DATETIME,        
    @fees DECIMAL(10, 2),        
    @discount DECIMAL(10, 2),        
    @months_duration INT,        
    @finalFees DECIMAL(10,2),        
    @registered_via_user INT,        
    @NewId INT OUTPUT        
AS        
BEGIN        
    SET NOCOUNT ON;      
  
    BEGIN TRY  
        BEGIN TRANSACTION;  
  
        DECLARE @member_id INT;  
        DECLARE @application_id INT;  
        DECLARE @createdMemberId INT;  
  
        -- Try to find the member by PersonId  
        SELECT @member_id = memberId   
        FROM Members   
        WHERE person_id = @PersonId;  
  
        -- If person is not a member, create a new one  
        IF @member_id IS NULL  
        BEGIN  
            EXEC prMember_Add @PersonId, @registered_via_user, @createdMemberId OUTPUT;  
  
            IF @createdMemberId = 0  
            BEGIN  
                -- Something went wrong in member creation  
                RAISERROR('Failed to create or find a valid member.', 16, 1);  
                ROLLBACK TRANSACTION;  
                RETURN;  
            END  
  
            SET @member_id = @createdMemberId;  
        END  
  
        -- Check if application already exists  
        SELECT @application_id = application_id
        FROM Applications  
        WHERE member_id = @member_id AND assignment_id = @assigment_Id;  
  
        IF @application_id IS NULL  
        BEGIN  
            -- Create new application  
            INSERT INTO Applications (member_id, assignment_id, created_date, application_status, last_update)  
            VALUES (@member_id, @assigment_Id, GETDATE(), 1, GETDATE());  
  
            SET @application_id = SCOPE_IDENTITY();  
        END  
        ELSE  
        BEGIN  
            -- Reactivate existing application  
            UPDATE Applications  
            SET application_status = 1,  
                last_update = GETDATE()  
            WHERE application_id = @application_id;  

			-- Mark the latest subscription for this application as renewed
			UPDATE Subscriptions
			SET isRenewed = 1
				WHERE subscription_id = (
			SELECT TOP 1 subscription_id
			FROM Subscriptions
			WHERE Application_id = @application_id
			  AND isDelete = 0
			  AND isRefund = 0
			ORDER BY subscriptions_end_date DESC
);
        END  
  
        -- Activate member  
        UPDATE Members  
        SET membership_status = 1  
        WHERE memberId = @member_id;  
  
        -- Insert new subscription  
        INSERT INTO Subscriptions (  
            Application_id,  
            subscriptions_start_date,  
            subscriptions_end_date,  
            fees,  
            Register_via_user,  
            months_duration,  
            discount,  
            final_fees,  
            isDelete,  
            isRefund  
        )  
        VALUES (  
            @application_id,  
            @subscription_start_date,  
            @subscription_end_date,  
            @fees,  
            @registered_via_user,  
            @months_duration,  
            @discount,  
            @finalFees,  
            0, -- isDelete  
            0  -- isRefund  
        );  
  
        SET @NewId = SCOPE_IDENTITY();  
  
        COMMIT TRANSACTION;  
    END TRY  
    BEGIN CATCH  
        ROLLBACK TRANSACTION;  
        SET @NewId = -1;  
        THROW;  
    END CATCH  
END  
GO
```

# [dbo].[prSubscription_Delete]

```sql
/****** Object:  StoredProcedure [dbo].[prSubscription_Delete]    Script Date: 8/1/2025 5:13:56 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[prSubscription_Delete]
@Subscription_Id int,
@RowAffected int output
AS
BEGIN
IF EXISTS (
    SELECT 1 FROM Subscriptions
    WHERE subscription_id = @Subscription_Id
      AND subscriptions_end_date <= GETDATE()
)
BEGIN
    RETURN -1; -- Don't allow deletion
END;

DECLARE @application_Id int=0;
SET @application_Id =
(
select Application_id from Subscriptions
WHERE subscription_id= 9
)

UPDATE Subscriptions
SET isDelete=1
Where subscription_id=@Subscription_Id
SET @RowAffected = @@ROWCOUNT;

-- Check if there's any other active subscription for the same application
Declare @IsActive int;
select @IsActive=1 from Subscriptions
WHERE GETDATE() BETWEEN subscriptions_start_date AND subscriptions_end_date
AND Application_id=@application_Id and isDelete = 0 and isRefund=0

-- If no active subscription remains, update the Applications table
if @IsActive is null
UPDATE Applications
SET application_status = 0
WHERE application_id=@application_Id

END;
GO
```

# [dbo].[prSubscription_Renew]

```sql
/****** Object:  StoredProcedure [dbo].[prSubscription_Renew]    Script Date: 8/1/2025 5:13:56 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[prSubscription_Renew]    
@Subscription_Id int,    
@MonthsDuration int,          
@Application_id int,          
@Subscription_Start_Date datetime,          
@Subscription_End_Date datetime,          
@Register_Via_User int,          
@Category_id int,        
@Discount float,        
@New_Subscription_Id int = 0 output         
AS  
BEGIN          
DECLARE @RenewSubs_Start_Date datetime          
DECLARE @RenewSubs_End_Date datetime          
DECLARE @CategoryFees decimal          
DECLARE @ApplicationStatus bit          
DECLARE @FinalFees decimal          
BEGIN TRY          
BEGIN TRANSACTION;          
      
--if user trying to renew a subscription with the id that already has a newer one → deny.      
IF EXISTS (      
    SELECT 1      
    FROM Subscriptions      
    WHERE subscription_id = @Subscription_Id    
      AND isDelete = 0      
      AND isRefund = 0     
   AND isRenewed =1    
)      
BEGIN          
    RETURN;    
END    
          
--get the category fees and the application status. I can send the fees from the subscription class but           
--this approach is better in case user change the fees of a category          
select @ApplicationStatus= Applications.application_status,@CategoryFees=Categories.fees from Applications          
INNER JOIN TrainerCategoryAssignments on TrainerCategoryAssignments.AssignmentId = Applications.assignment_id          
INNER JOIN Categories on Categories.category_id = TrainerCategoryAssignments.CategoryId          
Where application_id=@Application_id and Categories.category_id=@Category_id;          
          
--if application status was equal to 1 that mean the subscription still not finish yet. so i am going to add the subscription            
-- start date with the value of the previous subscription end date. otherwise I am going to add the current date to the subscription start dat          
if @ApplicationStatus=1      
BEGIN          
 SET @RenewSubs_Start_Date= @Subscription_End_Date      
END;          
ELSE          
BEGIN          
 SET @RenewSubs_Start_Date=getdate()          
END;          
          
--set the subscription end date. Add the months duration to start date to calculate the subscription end date.          
 SET @RenewSubs_End_Date=DATEADD(MONTH,@MonthsDuration,@RenewSubs_Start_Date)          
          
--calcualte finalFees.    
IF @Discount = 0    
    SET @FinalFees = @MonthsDuration * @CategoryFees;    
ELSE    
    SET @FinalFees = (@CategoryFees - (@CategoryFees * @Discount / 100))*@MonthsDuration;      
          
--add new subscription          
INSERT INTO Subscriptions(Application_id,subscriptions_start_date,subscriptions_end_date,fees,          
Register_via_user,months_duration,discount,final_fees,isDelete,isRefund)          
VALUES (@Application_id,@RenewSubs_Start_Date,@RenewSubs_End_Date,@CategoryFees,@Register_Via_User,@MonthsDuration,@Discount,@FinalFees,0,0)          
SET @New_Subscription_Id=SCOPE_IDENTITY()          
    
--update the field renew subscription, to Prevents multiple renewals on the same subscription.   
update Subscriptions    
Set isRenewed = 1    
where subscription_id=@Subscription_Id    
          
--update the application status. this status will be 0 when the subscription end date is come          
UPDATE Applications          
SET application_status = 1,        
last_update=GETDATE()        
         
Where application_id=@Application_id          
          
COMMIT TRANSACTION;          
    END TRY          
    BEGIN CATCH          
        ROLLBACK TRANSACTION;          
        THROW;          
    END CATCH          
END; 
GO
```

# [dbo].[prSubscription_UpdateSubscription]

```sql
/****** Object:  StoredProcedure [dbo].[prSubscription_UpdateSubscription]    Script Date: 8/1/2025 5:13:56 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[prSubscription_UpdateSubscription]
@subscription_Id int,
@subscription_Start_Date datetime,
@subscription_End_Date datetime,
@Fees decimal,
@Register_Via_User int,
@month_Duration int,
@Discount float
AS
BEGIN

DECLARE @final_Fees decimal;
IF @Discount = 0
 SET @final_Fees = @month_Duration*@Fees
ELSE
 SET @final_Fees = (@Fees-(@Fees*@Discount/100))*@month_Duration

UPDATE Subscriptions
SET subscriptions_start_date=@subscription_Start_Date, 
 subscriptions_end_date=@subscription_End_Date,
 Register_via_user=@Register_Via_User,
 @month_Duration=@month_Duration,
 discount=@Discount,
 final_fees=@final_Fees
 WHERE subscription_id=@subscription_Id and isRenewed=0-- and @subscription_End_Date>GETDATE()

 RETURN @@RowCount;
END;
GO
```

# [dbo].[prSubscriptions_getAllSubscribers]

```sql
/****** Object:  StoredProcedure [dbo].[prSubscriptions_getAllSubscribers]    Script Date: 8/1/2025 5:13:56 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[prSubscriptions_getAllSubscribers]
AS
BEGIN
WITH LatestSubscriptions AS (
    SELECT *,
           ROW_NUMBER() OVER (PARTITION BY Subscriptions.Application_id ORDER BY Subscriptions.subscription_id DESC) AS rn
    FROM Subscriptions
    WHERE isDelete = 0 AND isRefund = 0
)
SELECT 
    s.subscription_id AS [SUB ID],
    LTRIM(RTRIM(
        MemberPeople.first_name +
        CASE
            WHEN MemberPeople.second_name IS NULL OR MemberPeople.second_name = '' THEN ''
            ELSE ' ' + MemberPeople.second_name
        END +
        ' ' + MemberPeople.last_name
    )) AS [MEMBER NAME],
    TrainerPeople.first_name + ' ' + TrainerPeople.second_name + ' ' + TrainerPeople.last_name AS [TRAINER NAME],
    Categories.category_name AS [CATEGORY],
    CAST(s.subscriptions_start_date AS DATE) AS [START DATE],
    CAST(s.subscriptions_end_date AS DATE) AS [END DATE],
    CASE
        WHEN CAST(s.subscriptions_end_date AS DATE) >= CAST(GETDATE() AS DATE)
        THEN 'Active'
        ELSE 'InActive'
    END AS [STATUS],
    s.fees AS [FEES],
    s.discount,
    s.final_fees AS [FINAL FEES],
    Users.userName
FROM LatestSubscriptions s
INNER JOIN Applications ON Applications.application_id = s.Application_id
INNER JOIN Members ON Members.memberId = Applications.member_id
INNER JOIN People AS MemberPeople ON MemberPeople.personid = Members.person_Id
INNER JOIN TrainerCategoryAssignments ON TrainerCategoryAssignments.AssignmentId = Applications.assignment_id
INNER JOIN Trainers ON Trainers.trainer_id = TrainerCategoryAssignments.TrainerId
INNER JOIN People AS TrainerPeople ON TrainerPeople.personid = Trainers.person_Id
INNER JOIN Categories ON Categories.category_id = TrainerCategoryAssignments.CategoryId
INNER JOIN Users ON Users.userid = s.Register_via_user
WHERE s.rn = 1
ORDER BY s.subscription_id DESC;
END;
GO
```

# [dbo].[prTrainerCategory_Add]

```sql
/****** Object:  StoredProcedure [dbo].[prTrainerCategory_Add]    Script Date: 8/1/2025 5:13:56 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[prTrainerCategory_Add]  
    @TrainerId INT,  
    @CategoryId INT,  
    @Salary SMALLMONEY,  
    @NewId INT OUTPUT  
AS  
BEGIN  
    SET @NewId = 0;

    DECLARE @status INT;
    SET @status = dbo.fnCheckTrainerAssignmentStatus(@TrainerId, @CategoryId);

    IF @status = 1
        RETURN;

    IF @status = 2
    BEGIN
        UPDATE TrainerCategoryAssignments
        SET IsDelete = 0, Salary = @Salary
        WHERE TrainerId = @TrainerId AND CategoryId = @CategoryId;

        SELECT @NewId = AssignmentId 
        FROM TrainerCategoryAssignments 
        WHERE TrainerId = @TrainerId AND CategoryId = @CategoryId;

        RETURN;
    END

    -- Insert new assignment
    INSERT INTO TrainerCategoryAssignments(TrainerId, CategoryId, Salary, IsDelete)  
    VALUES (@TrainerId, @CategoryId, @Salary, 0);

    SET @NewId = SCOPE_IDENTITY();
END;
GO
```

# [dbo].[prUser_isExist]

```sql
/****** Object:  StoredProcedure [dbo].[prUser_isExist]    Script Date: 8/1/2025 5:13:56 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE procedure [dbo].[prUser_isExist]    
@UserName varchar(50),        
@Password varchar(100)          
AS          
BEGIN          
 Declare @IsExist Bit;          
 SELECT @IsExist = (CASE          
 WHEN Count(*) >0 Then 1           
 ELSE 0          
 END )          
 From Users          
 Where userName = @UserName and Password=@Password and IsDeleted=0;          
 return @IsExist;        
END;  
GO
```

# [dbo].[prUser_update]
